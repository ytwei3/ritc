# speedbump

```python
import requests

ticker = "DOVE"
key = "114514"

def get_speed_bump():
    with requests.Session() as s:
        s.headers.update({"X-API-Key": key})
        while True:
            r = s.post(
                "http://localhost:9999/v1/orders?type=MARKET&quantity=1&action=BUY&ticker=" + ticker
            )

            print(r.json())
            if r.status_code == 429:
                print("Too many requests, please wait...")

get_speed_bump()
```

# Latency

```python
import time
import requests

ticker = "ALPHA"
with requests.Session() as s:
    s.headers.update({"X-API-Key": "test"})
    s.post("http://localhost:9999/v1/orders?type=MARKET&quantity=1&action=BUY&ticker=" + ticker)

    start = time.time()
    for i in range(5):
        s.post("http://localhost:9999/v1/orders?type=MARKET&quantity=1&action=BUY&ticker=" + ticker)
    
    elapsed = time.time() - start
    print("get request avg. time: " + str(elapsed / 5) + " seconds")
```
